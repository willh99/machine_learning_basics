```python
import numpy as np
import matplotlib.pyplot as plt

# X represents the features of our training data, the diameters of pizzas
X = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)

# y is a vector representing the prices of the pizzas
y = [7, 9, 13, 17.5, 18]

plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(X, y, 'ko')
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Create an instance of the estimator, LinearRegression
model = LinearRegression()

# Fit the model on the training data
model.fit(X, y)

# Predict the price of a pizza with a diameter that has never been seen before
test_pizza = np.array([[12]])
predicted_price = model.predict(test_pizza)[0]

print('A 12" pizza should cost: $%.2f' % predicted_price)
```

```python
# Shows the line of best fit for the sums of the distance (area of squares) of the data points
# from that line
print('Residual sum of squares: %.2f' % np.mean((model.predict(X) - y) ** 2))
```

```python
# Solving Ordinary Least Squares for the function.
# Used to calculate variance of x
# Variance is the measure of how far values are spread out 
x_bar = X.mean()
print(x_bar)

# Note that we subtract one from the number of training instances when calculating the sample
# variance. This technique is called Bessel's correction. It corrects the bias in the estimation
#  of the population from the sample
variance = ((X - x_bar)**2).sum() / (X.shape[0] - 1)
print(variance)
```

```python
# Calculate variance of x using np.var this time
print(np.var(X, ddof=1))
```

```python
# Covariance is the meaure of how two variables change together (here x and y).
# [x+ : y+]=positive covariance, 
# [x+ : y-]=negative covariance, [No linear relation]=zero covariance.
# Switching y from a list to a NumPy ndarray which provides a method to calculate the sample mean
y = np.array([7, 9, 13, 17.5, 18])
y_bar = y.mean()

# Transpose X because both operands must be row vectors
covariance = np.multiply((X- x_bar).transpose(), y - y_bar).sum() / (X.shape[0] -1)
print(covariance)
print(np.cov(X.transpose(), y)[0][1])

# Remember: in y=mx+b, b = (cov(x,y) / var(x) )
# Once b is solved for, m can then be solved for using the coordinates of the centroid,
# the point that the model must pass through, as the values of x and y
```

```python
# Evaluating the model
# Try to determine r-squared (coefficient of determination) to check how
# close data are to the regression line
X_train = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y_train = [7, 9, 13, 17.5, 18]

X_test = np.array([8, 9, 11, 16, 12]).reshape(-1, 1)
y_test = [11, 8.5, 15, 18, 11]

model = LinearRegression()
model.fit(X_train, y_train)
r_sqr = model.score(X_test, y_test)
print(r_sqr)
```
